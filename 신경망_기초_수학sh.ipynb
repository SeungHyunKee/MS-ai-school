```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
# 일차함수 :   y = ax + b

def linear_function(x):
  # 기울기지정
  a = 0.5
  b = 2  # 일단이렇게 값 고정

  return a * x + b
```

```python
print(linear_function(5))
```

```python
# 선으로표시하기위해선 값이 여러개있어야함 :  x값을 배열로 만들어서 던진다

# x값 배열로 만들기 : 표현하려고 하는것 : -5 부터 5 까지 0.1씩
x = np.arange(-5,5,0.1)
y = linear_function(x)  # 이렇게 배열인 x값을 함수에 주면 배열에대한 값들 만큼 y에 들어온다

# 시각화하기
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
# 이차함수 ,  y = ax제곱 + bx + c

def quadratic_function(x):
  a = 1
  b = -1
  c = -2

  return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
y = quadratic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python
# 삼차함수(다항함수) ,  y = ax세제곱 + bx제곱 + cx + d

def cubic_function(x):
   a = 4
   b = 0
   c = -1
   d = -8

   return a*x**3 + b*x**2 + c*x + d
```

```python
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python
# 최소값을 만들기위한 함수 ? ?

def my_func(x):
  a=1
  b=-3
  c=10
  
  return a*x**2 + b*x + c
```

```python
x = np.arange(-10, 10, 0.1)
y = my_func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('my Function')

# minimum (지정해둔값으로)값 구하기
plt.scatter(1.5, my_func(1.5))  
```

```python
# 지수함수 : y = ax
def exponential_function(x):
  a = 4
  return a**x
```

```python
print(exponential_function(4))
print(exponential_function(0))
```

```python
# 지수성장의 다른표현방법

x = np.arange(-3, 2, 0.1)
y = exponential_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-4, 3)
plt.ylim(-1, 15)
plt.title('exponetial_function')
```
